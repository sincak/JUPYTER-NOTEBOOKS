# Exploring Beat Frequencies using the `Audio` Object

This example uses the `Audio` object and Matplotlib to explore the phenomenon of beat frequencies.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from ipywidgets import interactive
from IPython.display import Audio, display
import numpy as np
```

```python
def beat_freq(f1=220.0, f2=330.0,f3=440, f4=660,f5=440,f6=330):
    max_time = 3
    rate = 8000
    times = np.linspace(0,max_time,rate*max_time)
    signal = np.sin(2*np.pi*f1*times) + np.sin(2*np.pi*f2*times) + np.sin(2*np.pi*f3*times) + np.sin(2*np.pi*f4*times) + np.sin(2*np.pi*f5*times) + np.sin(2*np.pi*f6*times)
    display(Audio(data=signal, rate=rate))
    return signal
```

```python
v = interactive(beat_freq, f1=(200.0,800.0), f2=(200.0,800.0), f3=(200.0,800.0), f4=(200.0,800.0), f5=(200.0,800.0), f6=(200.0,800.0))
display(v)
```
