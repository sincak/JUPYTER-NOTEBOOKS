```
# @title
import numpy as np
from numpy import *
from math import *
from random import *
A=2;B=4;C=8
while np.gcd.reduce(array([A,B,C])) != 1:
  a=randint(1,1)
  b=randint(-6,6)
  if (b==-1 or b==0 or b==1): b=randint(2,8) #For Level 1 problems
  c=randint(1,1)
  d=randint(-6,6)
  if d==0: d=randint(-12,-1)
  A = a * c;B=a * d + b * c;C=b * d
  if np.gcd.reduce(array([A,B,C])) == 1: break

# And now MAJOR nesting to get the linear term right (sign, and not to show term when 0, and not to show coefficient when 1), and constant term sign!
if B == 0: show_B=""
else:
  if B < 0:
    if B == -1: show_B=f" - x "
    else: show_B=f" - {abs(B)}x"
  else:
    if B == 1: show_B=f" + x "
    else: show_B=f" + {abs(B)}x"
if C > 0: C_sign='+'
else: C_sign='-'

# And now for the factored forms...
if b > 0: b_sign='+'
else: b_sign ='-'
if d > 0: d_sign='+'
else: d_sign ='-'
babs=abs(b)
dabs=abs(d)
```
